## 🔍 Ejercicio 2 — `escaner()`: recorrer y clasificar archivos
⏱ *35 minutos*

### El problema

La estructura real del disco tiene carpetas anidadas, conductores como carpetas, y subcarpetas excluidas. El escáner debe:

1. Encontrar todos los `.mp3` y `.mp4`
2. Ignorar los que están en carpetas excluidas
3. Identificar a qué programa pertenece cada archivo subiendo por los ancestros
4. Manejar el caso donde la carpeta padre es un conductor

### ¿Cómo sabemos hasta dónde subir?

`archivo.relative_to(ruta_programas)` devuelve solo lo que está dentro de `PROGRAMAS/`. Cuando `ruta_actual.name == ""` llegamos al límite — ya no hay más carpetas dentro de la raíz.

```
PROGRAMAS/CESAR PAGANO/PAGANO BOLERO/audio.mp3
          ↑ conductor   ↑ programa    ↑ archivo

ruta_relativa.parts = ('CESAR PAGANO', 'PAGANO BOLERO', 'audio.mp3')
directorio_padre    = 'PAGANO BOLERO'  → en CARPETAS_PROGRAMAS ✅
```

### 2a — Versión base construida en clase

Analiza la salida. Identifica qué casos resuelve y cuáles no.

````python
from pathlib import Path

def escaner_base(ruta_programas: str):
    ruta_programas = Path(ruta_programas)

    for archivo in ruta_programas.rglob("*.mp3"):
        ruta_relativa    = archivo.relative_to(ruta_programas)
        directorio_padre = ruta_relativa.parent.name

        if directorio_padre in CARPETAS_EXCLUIDAS:
            continue
        elif directorio_padre in CARPETAS_PROGRAMAS:
            print(f"✅ PROGRAMA  : {directorio_padre} → {archivo.name}")
        elif directorio_padre in CONDUCTORES:
            if ruta_relativa.parent.parent.name == "":
                print(f"⚠️  CONDUCTOR sin programa: {directorio_padre} → {archivo.name}")
            else:
                if ruta_relativa.parent.parent.name in CARPETAS_PROGRAMAS:
                    print(f"✅ CONDUCTOR : {directorio_padre} → {archivo.name}")
        else:
            print(f"❓ NO ENCONTRADO: {ruta_relativa}")

print("── Salida del escáner base ──────────────────────────────")
escaner_base("/content/disco_prueba/PROGRAMAS")
````

### 2b — Ejercicio semi-guiado: versión completa con `.mp4` y `while`

Completa las partes marcadas con `# ← TU CÓDIGO`.

| Caso | `directorio_padre` | Solución |
|------|-------------------|---------|
| `PROGRAMA/audio.mp3` | en `CARPETAS_PROGRAMAS` | directo |
| `EXCLUIDA/audio.mp3` | en `CARPETAS_EXCLUIDAS` | `continue` |
| `CONDUCTOR/audio.mp3` | en `CONDUCTORES`, abuelo vacío | sin programa |
| `CONDUCTOR/PROGRAMA/audio.mp3` | en `CARPETAS_PROGRAMAS` | el padre ya es programa |
| `CONDUCTOR/SUBPROG/audio.mp3` | no reconocida | `while` sube hasta encontrar |

````python
from pathlib import Path

def escaner(ruta_programas: str):
    ruta_programas = Path(ruta_programas)
    resultados     = []

    # ← TU CÓDIGO: coleccionar archivos de TODAS las extensiones en EXTENSIONES_AUDIO
    # Pista: itera EXTENSIONES_AUDIO y usa rglob(f"*{ext}") en cada iteración
    archivos_encontrados = []
    for extension in EXTENSIONES_AUDIO:
        archivos_encontrados.extend(ruta_programas.rglob(f"*{extension}"))

    for archivo in archivos_encontrados:
        ruta_relativa    = archivo.relative_to(ruta_programas)
        directorio_padre = ruta_relativa.parent.name

        # ── Caso 1: carpeta excluida ─────────────────────────
        if directorio_padre in CARPETAS_EXCLUIDAS:
            continue

        # ── Caso 2: padre directo es el programa ─────────────
        elif directorio_padre in CARPETAS_PROGRAMAS:
            resultados.append({
                "programa":  directorio_padre,
                "conductor": None,
                "archivo":   archivo.name,
                "ruta":      archivo,
            })
            print(f"✅ PROGRAMA  : {directorio_padre} → {archivo.name}")

        # ── Caso 3: padre es un conductor ────────────────────
        elif directorio_padre in CONDUCTORES:
            ruta_actual         = ruta_relativa.parent.parent
            programa_encontrado = None

            while ruta_actual.name != "":
                if ruta_actual.name in CARPETAS_EXCLUIDAS:
                    break
                if ruta_actual.name in CARPETAS_PROGRAMAS:
                    programa_encontrado = ruta_actual.name
                    break
                ruta_actual = ruta_actual.parent

            if programa_encontrado:
                resultados.append({
                    "programa":  programa_encontrado,
                    "conductor": directorio_padre,
                    "archivo":   archivo.name,
                    "ruta":      archivo,
                })
                print(f"✅ CONDUCTOR : {directorio_padre} "
                      f"→ programa: {programa_encontrado} → {archivo.name}")
            else:
                print(f"⚠️  CONDUCTOR sin programa: {directorio_padre} → {archivo.name}")

        # ── Caso 4: carpeta no reconocida — subir a buscar ───
        else:
            # ← TU CÓDIGO:
            # Sube por los ancestros desde ruta_relativa.parent
            # Busca el programa y el conductor en el camino
            # Si encuentras el programa → append a resultados
            # Si no → imprime el error
            ruta_actual          = ruta_relativa.parent
            programa_encontrado  = None
            conductor_encontrado = None

            while ruta_actual.name != "":
                if ruta_actual.name in CARPETAS_EXCLUIDAS:
                    break
                if ruta_actual.name in CONDUCTORES and conductor_encontrado is None:
                    conductor_encontrado = ruta_actual.name
                if ruta_actual.name in CARPETAS_PROGRAMAS:
                    programa_encontrado = ruta_actual.name
                    break
                ruta_actual = ruta_actual.parent

            if programa_encontrado:
                resultados.append({
                    "programa":  programa_encontrado,
                    "conductor": conductor_encontrado,
                    "archivo":   archivo.name,
                    "ruta":      archivo,
                })
                print(f"✅ SUBCARPETA: {directorio_padre} "
                      f"→ programa: {programa_encontrado} → {archivo.name}")
            else:
                print(f"❌ SIN PROGRAMA: {ruta_relativa}")

    return resultados

print("── Salida del escáner completo ──────────────────────────")
resultados = escaner("/content/disco_prueba/PROGRAMAS")
print()
print(f"Total archivos catalogables: {len(resultados)}")
````

## 🧱 Ejercicio 3 — `construir_registro()`: un archivo → una fila
⏱ *20 minutos*

Cada archivo encontrado por `escaner()` debe convertirse en un **registro del catálogo** — una fila del DataFrame con todas sus columnas.

### Metadatos que leemos con `pathlib`

```python
path = Path("audio.mp3")

path.name                # "audio.mp3"          nombre original
path.suffix              # ".mp3"               extensión
path.stat().st_size      # 29768192             tamaño en bytes → dividir entre 1024*1024 para MB
path.stat().st_mtime     # 1699622400.0         timestamp de modificación
str(path.resolve())      # "/ruta/completa/..."  ruta absoluta
```

Convertir timestamp a fecha legible:
```python
from datetime import datetime
datetime.fromtimestamp(path.stat().st_mtime).strftime("%Y-%m-%d %H:%M:%S")
```

````python
from pathlib import Path
from datetime import datetime

def construir_registro(programa, conductor, ruta_archivo):
    path      = Path(ruta_archivo)
    stat      = path.stat()
    tamaño_mb = round(stat.st_size / (1024 * 1024), 4)
    fecha     = datetime.fromtimestamp(stat.st_mtime).strftime("%Y-%m-%d %H:%M:%S")

    return {
        "nombre_programa":            programa,
        "conductor":                  conductor or "",
        "nombre_archivo_original":    path.name,
        "nombre_archivo_normalizado": "",       # se completará en S6
        "tipo_archivo":               path.suffix.lower(),
        "tamaño_mb":                  tamaño_mb,
        "ruta_sistema":               str(path.resolve()),
        "fecha_modificacion":         fecha,
        "tiene_backup":               False,
    }

# ── Prueba con el primer resultado del escáner ────────────
if resultados:
    r0      = resultados[0]
    registro = construir_registro(r0["programa"], r0["conductor"], r0["ruta"])
    print("🧪 Registro de prueba:")
    for clave, valor in registro.items():
        print(f"   {clave:35s}: {repr(valor)}")
````

## 📋 Ejercicio 4 — `construir_catalogo()`: armar el DataFrame completo
⏱ *20 minutos*

### Crear un DataFrame desde una lista de diccionarios

```python
import pandas as pd

filas = [
    {"nombre": "audio1.mp3", "programa": "BARRIO ADENTRO"},
    {"nombre": "audio2.mp3", "programa": "EN LA RAYA"},
]

df = pd.DataFrame(filas)
# cada diccionario → una fila
# cada clave       → una columna
```

````python
import pandas as pd

def construir_catalogo(ruta_programas: str):
    archivos_encontrados = escaner(ruta_programas)

    if not archivos_encontrados:
        print("⚠️  No se encontraron archivos de audio")
        return pd.DataFrame()

    filas = []
    for resultado in archivos_encontrados:
        registro = construir_registro(
            resultado["programa"],
            resultado["conductor"],
            resultado["ruta"],
        )
        filas.append(registro)

    return pd.DataFrame(filas)

print("🔍 Construyendo catálogo...")
print()
df_catalogo = construir_catalogo("/content/disco_prueba/PROGRAMAS")
print()
print(f"✅ {len(df_catalogo)} registros | {len(df_catalogo.columns)} columnas")
````

## 💾 Ejercicio 5 — `guardar_catalogo()`: guardar como CSV
⏱ *15 minutos*

### `utf-8-sig` vs `utf-8`

```python
df.to_csv("catalogo.csv", encoding="utf-8")      # tildes ok en Python
df.to_csv("catalogo.csv", encoding="utf-8-sig")  # tildes ok en Excel también
```

`utf-8-sig` agrega una marca invisible al inicio (BOM) que Excel usa para leer caracteres con tilde. Sin esto, "BARRIO ADENTRO" puede verse como texto roto en Excel.

### `na_rep=""`

```python
df.to_csv("catalogo.csv", na_rep="")   # None/NaN → celda vacía ✅
df.to_csv("catalogo.csv")              # None/NaN → el texto "NaN"  ❌
```

````python
def guardar_catalogo(df, ruta_csv="catalogo_oriente.csv"):
    try:
        df.to_csv(
            ruta_csv,
            index=False,
            encoding="utf-8-sig",
            na_rep="",
        )
        tamaño_kb = Path(ruta_csv).stat().st_size / 1024
        print(f"✅ Catálogo guardado: {ruta_csv}")
        print(f"   {len(df)} registros · {tamaño_kb:.1f} KB")
        return True
    except PermissionError:
        print(f"❌ Sin permisos para escribir en: {ruta_csv}")
        return False
    except Exception as e:
        print(f"❌ Error al guardar: {e}")
        return False

guardar_catalogo(df_catalogo)
````

## 🔁 Ejercicio 6 — Iterar el DataFrame: leer, filtrar y consultar
⏱ *25 minutos*

### Tres formas de recorrer un DataFrame

```python
# Forma 1: columna completa (para cálculos sobre toda la columna)
for valor in df["nombre_programa"]:
    print(valor)

# Forma 2: iterrows() — fila por fila, decisión por archivo
for indice, fila in df.iterrows():
    print(fila["nombre_programa"], fila["tipo_archivo"])

# Forma 3: filtros vectorizados — la más eficiente con pandas
df_filtrado = df[df["tipo_archivo"] == ".mp3"]
```

Usamos `iterrows()` cuando necesitamos tomar una **decisión por archivo**.
Usamos filtros cuando queremos **consultas generales** sobre el catálogo.

````python
import pandas as pd
df = pd.read_csv("catalogo_oriente.csv", encoding="utf-8-sig", keep_default_na=False)

print("=" * 55)
print("📊 RESUMEN DEL CATÁLOGO — ORIENTE STEREO")
print("=" * 55)
print(f"  Total registros        : {len(df)}")
print(f"  Programas distintos    : {df['nombre_programa'].nunique()}")
print(f"  Con backup             : {(df['tiene_backup'] == True).sum()}")
print(f"  Sin backup             : {(df['tiene_backup'] == False).sum()}")
print(f"  Tamaño total (MB)      : {df['tamaño_mb'].sum():.4f}")
````

### Detectar duplicados por `ruta_sistema`

Dos registros son el mismo archivo si tienen la misma `ruta_sistema`.

```python
rutas_existentes = set(df_existente["ruta_sistema"].tolist())

if str(nueva_ruta.resolve()) not in rutas_existentes:
    # es nuevo → agregar
```

````python
import pandas as pd
from pathlib import Path

def actualizar_catalogo(ruta_programas: str, ruta_csv="catalogo_oriente.csv"):
    # Paso 1: cargar catálogo existente
    if Path(ruta_csv).exists():
        df_existente     = pd.read_csv(ruta_csv, encoding="utf-8-sig", keep_default_na=False)
        rutas_existentes = set(df_existente["ruta_sistema"].tolist())
        print(f"📂 Catálogo existente: {len(df_existente)} registros")
    else:
        df_existente     = pd.DataFrame()
        rutas_existentes = set()
        print("📂 Catálogo vacío — se creará desde cero")

    # Paso 2: escanear y encontrar archivos nuevos
    archivos_encontrados = escaner(ruta_programas)
    nuevos_registros     = []

    for resultado in archivos_encontrados:
        ruta_str = str(resultado["ruta"].resolve())
        if ruta_str not in rutas_existentes:
            registro = construir_registro(
                resultado["programa"],
                resultado["conductor"],
                resultado["ruta"],
            )
            nuevos_registros.append(registro)

    # Paso 3: unir y guardar
    if nuevos_registros:
        df_nuevos = pd.DataFrame(nuevos_registros)
        if df_existente.empty:
            df_actualizado = df_nuevos
        else:
            df_actualizado = pd.concat([df_existente, df_nuevos], ignore_index=True)
        guardar_catalogo(df_actualizado, ruta_csv)
        print(f"✅ {len(nuevos_registros)} archivos nuevos agregados")
        return df_actualizado
    else:
        print("✅ No hay archivos nuevos — el catálogo está al día")
        return df_existente

print("── Primera actualización ────────────────────────────────")
df_actualizado = actualizar_catalogo("/content/disco_prueba/PROGRAMAS")
````
